# PART-II : GDP and Education Drop-out Rates

```python
df_dropout = pd.read_csv('Dropouts.csv')
df_dropout.head()
```

```python
df_dropout = df_dropout.rename(columns = {'Primary - 2014-2015':'Primary - 2013-2014','Primary - 2014-2015.1':'Primary - 2014-2015','Level of Education - State':'State'})
df_dropout
```

```python
df_dropout = df_dropout[['State','Primary - 2014-2015','Upper Primary - 2014-2015','Secondary - 2014-2015']]

df_dropout.head()
```

```python
#Dropping the states having null values
df_dropout = df_dropout.dropna(how='any')

df_dropout
```

```python
df_dropout = df_dropout.replace(['Chhatisgarh','Uttrakhand'],['Chhattisgarh','Uttarakhand'])

df_dropout
```

```python
df_drpcap = pd.merge(df_all_states[df_all_states.Item=='Per Capita GSDP (Rs.)'], df_dropout, how = 'inner', on = 'State')

df_drpcap
```

```python
df_drpcap['Total_dropout_in_2014-15'] = df_drpcap.iloc[:,-3:].sum(axis = 1)

df_drpcap
```

```python
x = df_drpcap['2014-15'].values 

y1 = df_drpcap['Primary - 2014-2015'].values 

y2 = df_drpcap['Upper Primary - 2014-2015'].values 

y3 = df_drpcap['Secondary - 2014-2015'].values 

y4 = df_drpcap['Total_dropout_in_2014-15'].values 

plt.figure(figsize=(14,12))

plt.subplot(221)

plt.title('Primary')

plt.xlabel('GSDP in Crores (Rs.)')

plt.ylabel('Dropout rate in Percentage')

plt.scatter(x,y1)

plt.subplot(222)

plt.title('Upper-Primary')

plt.xlabel('GSDP in Crores (Rs.)')

plt.ylabel('Dropout rate in Percentage')

plt.scatter(x,y2)

plt.subplot(223)

plt.title('Secondary')

plt.xlabel('GSDP in Crores (Rs.)')

plt.ylabel('Dropout rate in Percentage')

plt.scatter(x,y3)

plt.subplot(224)

plt.title('Total Dropout Rate')

plt.xlabel('GSDP in Crores (Rs.)')

plt.ylabel('Dropout rate in Percentage')

plt.scatter(x,y4)

plt.show()
```

The Scatter plot shows that there is an increase in GSDP with decrease in Dropout rate.
